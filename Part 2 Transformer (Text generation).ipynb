```python
#1

import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Step 1: Load pre-trained model
model_name = 'gpt2'
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Add a padding token to the tokenizer
tokenizer.add_special_tokens({'pad_token': '[PAD]'})  # You can choose any string for padding token, '[PAD]' is commonly used

# Step 2: Fine-tune the pre-trained model on a customized dataset
# Assuming you have your own dataset and you want to fine-tune the model on it
# Replace 'custom_dataset.txt' with the path to your dataset
custom_dataset_path = '/content/file.txt'

# Load and preprocess the dataset
with open(custom_dataset_path, 'r', encoding='utf-8') as file:
    dataset_text = file.read()

# Tokenize the dataset
inputs = tokenizer(dataset_text, return_tensors="pt", max_length=512, truncation=True, padding=True)

# Fine-tune the model
model = GPT2LMHeadModel.from_pretrained(model_name)  # Re-initialize model
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
model.train()
for epoch in range(3):  # Adjust the number of epochs as needed
    optimizer.zero_grad()
    outputs = model(**inputs, labels=inputs['input_ids'])
    loss = outputs.loss
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}: Loss - {loss.item()}")

# Save the fine-tuned model
model.save_pretrained("fine_tuned_gpt2_model")
```

```python
#2

import torch
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load fine-tuned model and tokenizer
model_path = "fine_tuned_gpt2_model"  # Path to the directory where fine-tuned model is saved
# Save the fine-tuned model
model.save_pretrained("fine_tuned_gpt2_model")

# Save the tokenizer
tokenizer.save_pretrained("fine_tuned_gpt2_model")


# Given sentence
input_text = "Once upon a time, there was a king who ruled over a prosperous kingdom."

# Tokenize input
input_ids = tokenizer.encode(input_text, return_tensors='pt')

# Generate text
output = model.generate(input_ids, max_length=200, num_return_sequences=1, do_sample=True)

# Decode generated output
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)


```
